```python
df1 = pd.DataFrame({
    'Roll No.': [1, 2, 3, 4, 5],
    'Maths': [45, 78, 45, 90, 66],
    'Physics': [33, 67, 12, 90, 44]
})

df2 = pd.DataFrame({
    'Roll No.': [6, 7, 8, 9, 10],
    'Maths': [78, 73, 45, 90, 69],
    'Physics': [23, 67, 88, 0, 98]
})

print(df1)
print(df2)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

Merge()

```python
data1 = pd.DataFrame({
    'Roll No.': [1, 2, 3, 4, 5],
    'Maths': [45, 78, 45, 90, 66]})
print(data1)
data2 =  pd.DataFrame({
    'Roll No.': [1, 2, 3, 4, 5],
    'Physics': [56, 78, 23, 90, 34]})
print(data2)
```

```python
merged_data = pd.merge(data1, data2, on = 'Roll No.')
merged_data
```

```python
data3 = pd.DataFrame({
    'Roll No.': [1, 2, 3, 4, 5],
    'Maths': [45, 78, 45, 90, 66]})
print(data3)
data4 =  pd.DataFrame({
    'Roll No.': [1, 2, 3, 7,8],
    'Physics': [56, 78, 23, 90, 34]})
print(data4)
```

```python
merged_data = pd.merge(data3, data4)
merged_data
```

```python
merged_data = pd.merge(data3, data4, how = 'left')
merged_data
```

```python
merged_data = pd.merge(data3, data4, how = 'right')
merged_data
```

```python
merged_data = pd.merge(data3, data4, how = 'outer')
merged_data
```
